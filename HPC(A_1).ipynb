```python

#Section 3: Section C — Interactive Compute Session (salloc/srun or qsub -I)
import time
import sys
print("Interactive Compute Session Test")
print("Python Version:", sys.version)
n = 10_000_000
total = 0
start = time.time()
for i in range(n):
    total += i
end = time.time()
print("Sum:", total)
print("Execution Time:", end - start, "seconds")
```

```python
#Section 4: Section D — Job Arrays for Parameter Sweep (Multiple Serial Runs)
import os
import random
import time

task_id = os.getenv("SLURM_ARRAY_TASK_ID") or os.getenv("PBS_ARRAYID")
task_id = int(task_id) if task_id is not None else 0

samples_list = [1000000, 2000000, 5000000, 10000000]
num_samples = samples_list[task_id % len(samples_list)]

inside = 0
start = time.time()

for _ in range(num_samples):
    x = random.random()
    y = random.random()
    if x * x + y * y <= 1.0:
        inside += 1

pi = 4 * inside / num_samples
end = time.time()

print("Task ID:", task_id)
print("Samples:", num_samples)
print("Estimated Pi:", pi)
print("Execution Time:", end - start)
```

```python
#Section 5: Section E — Modules/Venv, Scratch I/O, and Resource Flags
import os
import time
import sys
scratch_dir = os.environ.get("SLURM_TMPDIR", f"/scratch/{os.environ.get('USER', 'user')}")
work_dir = os.path.join(scratch_dir, "section5_job")
os.makedirs(work_dir, exist_ok=True)
output_file = os.path.join(work_dir, "output.txt")
with open(output_file, "w") as f:
    f.write("Section 5: Serial Python Job\n")
    f.write(f"Python Version: {sys.version}\n")
    f.write("Writing output to scratch directory\n\n")
    total = 0
    for i in range(1_000_000):
        total += i
    f.write(f"Computation Result: {total}\n")
time.sleep(1)
print("Job completed successfully")
print("Output written to:", output_file)
```

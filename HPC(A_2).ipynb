```python
#Q4: Monte Carlo Ï€ Estimation (Random + Reduction)
import random
import time
import cProfile
import pstats
import io
import tracemalloc
def monte_carlo_pi(num_samples):
    inside_circle = 0
    for _ in range(num_samples):
        x = random.random()
        y = random.random()
        if x * x + y * y <= 1.0:
            inside_circle += 1
    return 4 * inside_circle / num_samples
num_samples = 5_000_000
tracemalloc.start()
pr = cProfile.Profile()
pr.enable()
start_time = time.time()
pi_estimate = monte_carlo_pi(num_samples)
end_time = time.time()
pr.disable()
current, peak = tracemalloc.get_traced_memory()
tracemalloc.stop()
print("Estimated Pi:", pi_estimate)
print("Execution Time:", end_time - start_time, "seconds")
print(f"Current Memory Usage: {current / 10**6:.2f} MB")
print(f"Peak Memory Usage: {peak / 10**6:.2f} MB")
s = io.StringIO()
ps = pstats.Stats(pr, stream=s).sort_stats("cumulative")
ps.print_stats(10)
print(s.getvalue())
```

```python
#5th question
import random
import time
import cProfile
import pstats
import io
import tracemalloc
def generate_positions(n):
    return [(random.random(), random.random(), random.random()) for _ in range(n)]
def pairwise_kernel(positions):
    n = len(positions)
    total = 0.0
    for i in range(n):
        xi, yi, zi = positions[i]
        for j in range(i + 1, n):
            xj, yj, zj = positions[j]
            dx = xi - xj
            dy = yi - yj
            dz = zi - zj
            total += dx * dx + dy * dy + dz * dz
    return total
def run_and_measure(n):
    positions = generate_positions(n)

    tracemalloc.start()
    start = time.perf_counter()
    pairwise_kernel(positions)
    elapsed = time.perf_counter() - start
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return elapsed, peak
print("N\tTime (s)\tPeak Memory (KB)")
print("-" * 40)
for n in [200, 400, 800, 1200]:
    t, mem = run_and_measure(n)
    print(f"{n}\t{t:.4f}\t\t{mem / 1024:.1f}")
print("\n=== cProfile Hotspot Analysis ===")
positions = generate_positions(600)
profiler = cProfile.Profile()
profiler.enable()
pairwise_kernel(positions)
profiler.disable()
stream = io.StringIO()
stats = pstats.Stats(profiler, stream=stream)
stats.strip_dirs().sort_stats("cumulative").print_stats(10)
print(stream.getvalue())
```

```python
#Q1
import time
import random
import cProfile
import pstats
import io
import tracemalloc
N = 1_000_000
def generate_vectors(n):
    a = [random.random() for _ in range(n)]
    b = [random.random() for _ in range(n)]
    return a, b
def vector_add(a, b):
    result = []
    for i in range(len(a)):
        result.append(a[i] + b[i])
    return result
def dot_product(a, b):
    total = 0.0
    for i in range(len(a)):
        total += a[i] * b[i]
    return total
def main():
    a, b = generate_vectors(N)
    start = time.time()
    c = vector_add(a, b)
    add_time = time.time() - start

    start = time.time()
    dp = dot_product(a, b)
    dot_time = time.time() - start

    print(f"Vector Addition Time: {add_time:.4f} seconds")
    print(f"Dot Product Time: {dot_time:.4f} seconds")
    print(f"Dot Product Result: {dp:.4f}")

if __name__ == "__main__":
    tracemalloc.start()
    profiler = cProfile.Profile()
    profiler.enable()
    main()
    profiler.disable()
    s = io.StringIO()
    stats = pstats.Stats(profiler, stream=s).sort_stats("cumulative")
    stats.print_stats(10)
    print(s.getvalue())
    current, peak = tracemalloc.get_traced_memory()
    print(f"Current Memory Usage: {current / 10**6:.2f} MB")
    print(f"Peak Memory Usage: {peak / 10**6:.2f} MB")
    tracemalloc.stop()
```

```python

#Question 2
import time
import random
import cProfile
import pstats
import io
import tracemalloc
def generate_matrix(n):
    matrix = []
    for i in range(n):
        row = []
        for j in range(n):
            row.append(random.random())
        matrix.append(row)
    return matrix
def matrix_multiply(A, B):
    n = len(A)
    C = []
    for i in range(n):
        C.append([0.0] * n)
    for i in range(n):
        for j in range(n):
            total = 0.0
            for k in range(n):
                total += A[i][k] * B[k][j]
            C[i][j] = total
    return C
def run_benchmark(n):
    A = generate_matrix(n)
    B = generate_matrix(n)
    start = time.perf_counter()
    C = matrix_multiply(A, B);
    elapsed = time.perf_counter() - start

    return elapsed, C
if __name__ == "__main__":
    N = 150
    tracemalloc.start()
    profiler = cProfile.Profile()
    profiler.enable()

    elapsed_time, result_matrix = run_benchmark(N)

    profiler.disable()
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()

    print(f"Matrix size: {N} x {N}")
    print(f"Execution time: {elapsed_time:.4f} seconds")
    print(f"Current memory usage: {current / 1e6:.2f} MB")
    print(f"Peak memory usage: {peak / 1e6:.2f} MB")
    s = io.StringIO()
    stats = pstats.Stats(profiler, stream=s).sort_stats("cumtime")
    stats.print_stats(10)
    print("\n--- cProfile Top 10 Functions ---")
    print(s.getvalue())
```

```python
#Section 3: 2D Convolution (Stencil/Blur) on a Grid
import time
import random
import cProfile
import pstats
import io
import tracemalloc
ROWS = 200
COLS = 200
KERNEL_SIZE = 5
OFFSET = KERNEL_SIZE // 2
def generate_image(rows, cols):
    return [[random.random() for _ in range(cols)] for _ in range(rows)]
def blur_5x5(image, rows, cols):
    output = [[0.0 for _ in range(cols)] for _ in range(rows)]

    for i in range(OFFSET, rows - OFFSET):
        for j in range(OFFSET, cols - OFFSET):
            total = 0.0
            for ki in range(-OFFSET, OFFSET + 1):
                for kj in range(-OFFSET, OFFSET + 1):
                    total += image[i + ki][j + kj]
            output[i][j] = total / (KERNEL_SIZE * KERNEL_SIZE)
    return output
def main():
    image = generate_image(ROWS, COLS)
    print(f"Matrix Size: {ROWS} x {COLS}")
    start = time.time()
    blurred_image = blur_5x5(image, ROWS, COLS)
    elapsed = time.time() - start
    print(f"Blur Execution Time: {elapsed:.4f} seconds")
if __name__ == "__main__":
    tracemalloc.start()
    profiler = cProfile.Profile()
    profiler.enable()
    main()
    profiler.disable()
    s = io.StringIO()
    stats = pstats.Stats(profiler, stream=s).sort_stats("cumulative")
    stats.print_stats(10)
    print(s.getvalue())
    current, peak = tracemalloc.get_traced_memory()
    print(f"Current Memory Usage: {current / 10**6:.2f} MB")
    print(f"Peak Memory Usage: {peak / 10**6:.2f} MB")
    tracemalloc.stop()
```
